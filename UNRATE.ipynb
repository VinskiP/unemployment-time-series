## Unrate data

```r
library("ggplot2")
library("readxl")

options(repr.plot.width=20, repr.plot.height=12)
```

```r
data = read_xls("~/Documents/Datasets/unrate.xls")
data = data.frame(data)

head(data)
```

```r
pvm = as.Date(data$DATE)
```

```r
unrate= ggplot(data, aes(x=pvm)) +
        geom_line(aes(y=UNRATEW, col="UNRATEW")) +
        geom_line(aes(y=UNRATEM, col="UNRATEM")) +
        geom_line(aes(y=UNRATEBLACK, col="UNRATEBLACK")) +
        geom_line(aes(y=UNRATELATINO, col="UNRATELATINO")) +
        xlab("") +
        ylab("Unemployment rate") +
        scale_y_continuous(labels=scales::unit_format(unit="%"))+
        theme_linedraw()


unrate
```
